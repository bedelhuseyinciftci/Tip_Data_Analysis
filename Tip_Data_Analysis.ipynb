```python
#1. kütüphaneleri yükle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
print(sns.get_dataset_names())
```

```python
bahsis_veri=sns.load_dataset('tips')
print(bahsis_veri)
```

```python
print(type(bahsis_veri))
```

```python
#column sözlükte anahtar olarak eski kolon adları değer olarak yeni kolon anahtarları verilir.
#inplace=True ile değişikliklerin veri çerçevesinde kalıcı olması sağlanır.
bahsis_veri.rename(columns={'total_bill'            : 'fatura_toplami',
                               'tip'                : 'bahsis',
                               'sex'                : 'cinsiyet',
                               'smoker'             : 'sigara',
                               'day'                : 'gun',
                               'time'               : 'zaman',
                               'size'               : 'kisi_sayisi'
                              }, inplace=True)
```

```python
print(bahsis_veri)
```

```python
print(bahsis_veri.info())
```

```python
print(bahsis_veri.describe())
```

```python
sns.scatterplot(x='fatura_toplami', y='bahsis', data=bahsis_veri)
```

```python
sns.regplot(x='fatura_toplami', y='bahsis',line_kws={'color':'red'}, data=bahsis_veri)
```

```python
sns.set_style('darkgrid') #grafik stili arka plan rengi
sns.set(rc={'figure.figsize':(11.7,8.27)}) #grafik boyutları
sns.scatterplot(x='fatura_toplami', y='bahsis', data=bahsis_veri, hue='cinsiyet',color='b',palette='Set1')
plt.xlabel('Fatura Toplamı')#x ekseni adı
plt.ylabel('Bahşiş')#y ekseni adı
plt.title('Fatura Toplamı ve Bahşiş')#başlık
plt.show()
```

```python
sns.relplot(x='fatura_toplami', y='bahsis', data=bahsis_veri, hue='sigara',style='cinsiyet',color='b',palette='Set1', height=10)
plt.xlabel('Fatura Toplamı')#x ekseni adı
plt.ylabel('Bahşiş')#y ekseni adı
plt.title('Fatura Toplamı ve Bahşiş')#başlık
plt.show()
```

```python
sns.relplot(x='fatura_toplami', y='bahsis', size='kisi_sayisi',sizes=(20,200), data=bahsis_veri, hue='sigara',style='cinsiyet',color='b',palette='Set1', height=10)
plt.xlabel('Fatura Toplamı')#x ekseni adı
plt.ylabel('Bahşiş')#y ekseni adı
plt.title('Fatura Toplamı ve Bahşiş')#başlık
plt.show()
```

```python
sns.relplot(x='fatura_toplami', y='bahsis', size='kisi_sayisi',col='zaman',sizes=(20,200), data=bahsis_veri, hue='sigara',style='cinsiyet',color='b',palette='Set1', height=10)
plt.xlabel('Fatura Toplamı')#x ekseni adı
plt.ylabel('Bahşiş')#y ekseni adı

plt.show()
```

```python
sns.relplot(x='fatura_toplami', y='bahsis', size='kisi_sayisi',col='cinsiyet',sizes=(20,200), data=bahsis_veri, hue='sigara',style='cinsiyet',color='b',palette='Set1', height=10)
plt.xlabel('Fatura Toplamı')#x ekseni adı
plt.ylabel('Bahşiş')#y ekseni adı

plt.show()
```
